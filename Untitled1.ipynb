```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Model performance data
model_data = {
    "Model Name": [
        "Ridge", "Linear Regression", "CatBoosting Regressor", "Random Forest Regressor",
        "AdaBoost Regressor", "XGBRegressor", "Lasso", "K-Neighbors Regressor", "Decision Tree Regressor"
    ],
    "R2 Score": [
        0.880593, 0.880433, 0.851632, 0.850506, 0.847300, 0.827797, 0.825320, 0.783813, 0.740136
    ]
}

df_models = pd.DataFrame(model_data)

# Plotting R² Score Comparison
plt.figure(figsize=(12, 6))
sns.barplot(x="R2 Score", y="Model Name", data=df_models, palette="crest")
plt.title("Model Comparison Based on R² Score", fontsize=14)
plt.xlabel("R² Score")
plt.ylabel("Model")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.xlim(0.7, 0.9)
plt.tight_layout()
plt.show()
```
